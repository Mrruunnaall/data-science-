```python
PERFORM DIMENSIONALITY REDUCTION USING PCA
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardizing the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Applying PCA
pca = PCA()
pca_data = pca.fit_transform(scaled_data)

# Explained variance (Scree plot)
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title('Cumulative Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()

# Reduce to 2 or 3 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])

print(pca_df.head())  # Display the PCA-transformed data
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler() #Standardize the features.
scaled_data = scaler.fit_transform(df.iloc[:, 1:])

pca = PCA() # Implementation PCA and determine the optimal number of principal components.
pca.fit(scaled_data)

explained_variance_ratio = pca.explained_variance_ratio_    # Explained variance ratio for each principal component.
#Explained variance ratio is a measure of the proportion of the total variance in the original dataset that is explained by each principal component.

plt.figure(figsize=(15, 10))
#plt.subplot(1, 2, 1)    # Scree plot
plt.plot(range(1, len(explained_variance_ratio) + 1),explained_variance_ratio, marker = 'o', linestyle = '--', color = 'r') #color = r- red.
plt.title('Scree Plot')
plt.xlabel('Principle Component')
plt.ylabel('Explained Variable Ratio')

#plt.tight_layout()
plt.show()
```

```python
Perform Hierarchical Clustering with Dendrogram
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Perform linkage for hierarchical clustering using Ward's method
linked = linkage(pca_data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for PCA Data (Wine Dataset)')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
CLUSTERING WITH ORIGINALL DATA
```

```python
# Apply K-Means Clustering.
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust the number of clusters
kmeans.fit(scaled_data)

labels = kmeans.labels_   # Cluster labels
```

```python
# Visualize Clustering.
import matplotlib.pyplot as plt
import seaborn as sns

pca_2d = PCA(n_components=2)  #Visualize the clusters in 2D using the first two principal components or using a scatterplot.
components_2d = pca_2d.fit_transform(scaled_data)

plt.figure(figsize=(15, 8))
sns.scatterplot(x=components_2d[:, 0], y=components_2d[:, 1], hue=labels, palette='Set1')
plt.title('K-Means Clustering Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

silhouette_avg = silhouette_score(scaled_data, labels)
davies_bouldin = davies_bouldin_score(scaled_data, labels)

print(f'Silhouette Score :{silhouette_avg}')
print(f'Davis-Bouldin Index : {davies_bouldin}')
```

```python
CLUSTERING WITH PCA DATA
```

```python
# Apply K-Means Clustering to PCA-Transformed Dataset.

scaler = StandardScaler() #Standardize the features.
scaled_data = scaler.fit_transform(df.iloc[:, 1:])

pca = PCA()   # Implementation PCA.
pca.fit(scaled_data)

n_components = 2  # number of components based on the explained variance (e.g., 95% variance)
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(scaled_data)

kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca.fit(pca_data)

labels_pca = kmeans_pca.labels_   # Cluster labels for PCA data
```

```python
# Visualize Clustering Results from PCA-Transformed Dataset.

plt.figure(figsize=(15, 8))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=labels_pca, palette='Set1')
plt.title('K-Means Clustering on PCA-Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Compare the silhouette score and Daviesâ€“Bouldin index of the clustering results from the original data and the PCA-transformed data.

silhouette_pca = silhouette_score(pca_data, labels_pca)
davies_bouldin_pca = davies_bouldin_score(pca_data, labels_pca)

print(f'Silhouette Score (Original Data): {silhouette_avg}')
print(f'Silhouette Score (PCA Data): {silhouette_pca}')

print(f'Daviesâ€“Bouldin Index (Original Data): {davies_bouldin}')
print(f'Daviesâ€“Bouldin Index (PCA Data): {davies_bouldin_pca}')
```

# 5: Comparison and Analysis:
**1.	Compare the clustering results obtained from the original dataset and PCA-transformed data.**

After performing clustering on both the original dataset and the PCA-transformed data, compares the results using metrics such as the silhouette score and Daviesâ€“Bouldin index. These metrics help to quantify the quality of the clusters:

**Silhouette Score:** A higher score indicates better separation between clusters. Comparing this score between the original and PCA-transformed data shows how well-defined the clusters are in each case.

**Daviesâ€“Bouldin Index:** A lower index value suggests better clustering. This metric helps to determine if the clusters are more compact and well-separated.

Additionally, visual comparisons of cluster assignments can be done by plotting the clusters for both datasets.

**2.	Discuss any similarities or differences observed in the clustering results.**

**Similarities:**

**Cluster Structure:** If both the original and PCA-transformed data maintain similar cluster structures, it suggests that PCA has preserved the essential relationships between features.                                                  
**Cluster Assign:** If many data points are assigned to the same clusters in both cases, PCA has not drastically altered the underlying data structure.

**Differences:**

**Cluster Boundaries:** PCA might alter the cluster boundaries due to the change in feature space, potentially leading to different cluster assignments.             
**Number of Significant Clusters:** In some cases, PCA might reveal a different number of meaningful clusters than the original data, especially if noise is reduced during dimensionality reduction.

**3.	Reflect on the impact of dimensionality reduction on clustering performance.**

**Improved Performance:**

**Noise Reduction:** PCA can reduce noise by focusing on the most significant components, which may lead to more distinct and compact clusters.                
**Computational Efficiency:** With fewer dimensions, clustering algorithms like K-means run faster, especially with large datasets.

**Potential Drawbacks:**

**Loss of Information:** If too many dimensions are reduced, important information might be lost, potentially leading to less meaningful clusters.
**Interpretability:** Clustering on PCA-transformed data might be harder to interpret because the principal components are linear combinations of the original features.


**4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.**

**When to Use PCA:**

**High-Dimensional Data:** When working with datasets with many features, PCA helps to reduce dimensionality, making clustering more manageable.               
**Correlated Features:** PCA is particularly beneficial when features are highly correlated, as it consolidates information into fewer, uncorrelated principal components.                                                            
**Noise Reduction:** PCA can enhance clustering by eliminating noise and focusing on the most significant patterns in the data.

**When to Avoid PCA:**

**Small Datasets:** In small datasets, the risk of losing crucial information due to dimensionality reduction is higher.                                       
**Interpretability:** If interpretability of the clusters in terms of the original features is crucial, clustering on the original data may be preferred.
